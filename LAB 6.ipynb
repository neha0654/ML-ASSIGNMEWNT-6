```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

feature1 = y[:, 0]
feature2 = y[:, 1]

# Create a Linear Regression model
model = LinearRegression()

# Reshape the independent feature to 2D
feature1 = feature1.reshape(-1, 1)

# Fit the model
model.fit(feature1, feature2)

# Make predictions
predicted_values = model.predict(feature1)

mse = mean_squared_error(feature2, predicted_values)
print("Mean Squared Error = ",mse)
```
